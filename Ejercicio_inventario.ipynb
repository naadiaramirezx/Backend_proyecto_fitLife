```
# --- Sistema de Inventario B谩sico ---

# El inventario principal de la tienda
# (producto: cantidad)
inventario = {
    "manzanas": 50,
    "pl谩tanos": 75,
    "naranjas": 30
}

# Precios por unidad
precios = {
    "manzanas": 1.5,
    "pl谩tanos": 0.75,
    "naranjas": 1.0
}


# Registro de ventas
ventas = {}
#Error 1: Error de L贸gica.
def procesar_venta(producto, cantidad):
  a = len(ventas)+1

  """
  Procesa la venta de un producto y actualiza el inventario.
  """
  print(f"\nIntentando vender {cantidad} de {producto}...")

  # Verifica si el producto existe
  if producto not in inventario:
    print(f"Error: Producto '{producto}' no existe en el inventario.")
    return

  stock_actual = inventario[producto]

  #Error 2: Error de L贸gica
  # Verifica si hay suficiente stock

  #ERROR LGICO: Se esta usando > en vez de >=.
  if stock_actual > cantidad:
    # Calcula el nuevo stock
    inventario[producto] = stock_actual - cantidad

    # Calcula el costo
    #ERROR LOGICO: Hace falta una validacion de si el producto contiene un precio.
    costo_total = precios[producto] * cantidad

    print(f"Venta exitosa. Costo total: ${costo_total}")
    print(f"Stock restante de {producto}: {inventario[producto]}")

    ventas.update({
      f"factura_{a}":{
        "producto":producto,
        "cantidad":cantidad,
        "precio_unitario":precios[producto],
        "total_venta":costo_total
        }
      })

  else:
    print(f"Error: Stock insuficiente. Solo quedan {stock_actual}")

#Error 3: Error en documentacion.
def agregar_stock(producto, cantidad, precio):
  """
  Agrega nuevo stock al inventario.
  """
  print(f"\nAgregando {cantidad} de {producto}...")

  if producto in inventario:
    inventario[producto] += cantidad
  else:
    #inventario[producto] = cantidad
    inventario.update({
        producto:cantidad
    })
    precios.update({
        producto:precio
    })

  print(f"Stock actualizado de {producto}: {inventario[producto]}")
#ERROR DE DOCUMENTACION: El comentario se帽ala un error de sintaxis inexistente en la funcion.
# Error de sintaxis en la definici贸n de esta funci贸n

def imprimir_reporte_inventario():
  """
  Imprime un reporte del estado actual del inventario.
  """
  print("\n--- REPORTE DE INVENTARIO ---")
  for producto, cantidad in inventario.items():
    print(f"- {producto.title()}: {cantidad} unidades")
  print("-----------------------------")

#Error 4: Error de L贸gica.
def generar_factura_global():
  """
  Imprime un reporte de las ventas globales.
  producto, cantidad, precio_unitario, total_venta
  """
  print("\n--- REPORTE DE VENTAS GLOBALES ---")
  #ERROR LOGICO: no se hace muestra de un mensaje cuando no hay ventas registradas.
  for x,obj in ventas.items():
    print(x, end=": ")
    for y in obj:
      print(f"{y}: {obj[y]}", end=", ")
    print(" ")

# --- Ejecuci贸n Principal del Programa ---

print("Bienvenido al sistema de inventario.")

# Caso 1: Venta exitosa
procesar_venta("manzanas", 10)

# Caso 2: Intento de agregar un producto nuevo (uvas)
# Nota: "uvas" no est谩 en la lista de 'precios'.
agregar_stock("uvas", 40, 1)

# Caso 3: Venta de un producto que no est谩 en 'precios'
procesar_venta("uvas", 15)

# Caso 4: Venta de un producto con stock en formato incorrecto
procesar_venta("naranjas", 5)
procesar_venta("naranjas", 1)

# Caso 5: Venta que excede el stock
procesar_venta("pl谩tanos", 100)

# Caso 6: Llamada a la funci贸n con error de sintaxis
imprimir_reporte_inventario()

# Caso 7: Llamada a una funci贸n que no existe
generar_factura_global()
```
